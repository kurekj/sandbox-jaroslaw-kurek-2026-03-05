```python
# calculate average price over each quarter
df["quarter"] = df["create_date"].dt.to_period("Q")
df["quarter"] = df["quarter"].dt.to_timestamp()

# plot average price over quarters
df.groupby("quarter")["price_m2"].mean().plot()
```

```python
# create a plot on number of properties over create_date, but bin it by month
df["create_date"].dt.to_period("M").value_counts().sort_index().plot()
```
